```python
training = pd.read_csv('training_set_data.csv')
pd.set_option('display.max_columns', 500)
training.head()
```

```python
plt.figure(figsize=(6,8))

sns.relplot(data=training , x= 'total_income', y= 'age',  
                hue='uninsured', palette="ch:r=-.5,l=.75")

plt.title('The relationship between insurance status, age and income', fontsize=17)
# plt.xlabel('New Customers', fontsize=15)
# plt.ylabel('Impressions', fontsize=15)

plt.show()

# sns.relplot(x="total_bill", y="tip", hue="smoker", data=tips);
```
