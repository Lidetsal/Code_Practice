hh_income_pct:
  description:
    - Percentage of the household's total income that comes from that person; 
      calculated as (person's total income) / (household income).

the total income earned per person; this can
be calculated, however, by multiplying the person’s percent of household
income (hh_income_pct) with the household income (hh_income) found in the
household-level data set.

```python
new = pd.merge(person, household, how='inner', on='household_id')

new.head()
```

```python
new['personal_income'] = new.hh_income_pct*new.hh_income

new = new.sort_values(by='personal_income', ascending=False)
new.head(3)
```
